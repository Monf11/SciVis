```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Supponendo che tu abbia due tabelle incrociate: worked e wanted
# worked = pd.crosstab(...)
# wanted = pd.crosstab(...)

# Somma delle tabelle incrociate
combined_cross_tab = worked + wanted

# Impostazione dei parametri
bar_width = 0.4
index = np.arange(len(combined_cross_tab.index))

# Creazione del grafico a barre combinato con indici verticali
fig, ax = plt.subplots(figsize=(12, 6))
bar1 = ax.bar(index - bar_width/2, worked.sum(axis=1), width=bar_width, label='Worked')
bar2 = ax.bar(index + bar_width/2, wanted.sum(axis=1), width=bar_width, label='Wanted')

ax.set_xlabel('Lingue')
ax.set_ylabel('Frequenza')
ax.set_title('Confronto tra "Worked" e "Wanted"')
ax.set_xticks(index)
ax.set_xticklabels(combined_cross_tab.index, rotation=90, ha='center')  # Ruota gli indici di 90 gradi
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Supponendo che tu abbia due tabelle incrociate: worked e wanted
# worked = pd.crosstab(...)
# wanted = pd.crosstab(...)

# Somma delle tabelle incrociate
combined_cross_tab = worked + wanted

# Impostazione dei parametri
bar_width = 0.6  # Aumentato
index = np.arange(len(combined_cross_tab.index))

# Creazione del grafico a barre combinato con indici verticali e dimensioni aumentate
fig, ax = plt.subplots(figsize=(16, 8))  # Aumentato
bar1 = ax.bar(index - bar_width/2, worked.sum(axis=1), width=bar_width, label='Worked')
bar2 = ax.bar(index + bar_width/2, wanted.sum(axis=1), width=bar_width, label='Wanted')

ax.set_xlabel('Lingue')
ax.set_ylabel('Frequenza')
ax.set_title('Confronto tra "Worked" e "Wanted"')
ax.set_xticks(index)
ax.set_xticklabels(combined_cross_tab.index, rotation=90, ha='center')  # Ruota gli indici di 90 gradi
ax.legend()

plt.show()
```

```python
df = pd.concat([worked, wanted], axis = 1,keys=['Worked', 'Wanted'])
df = df[df['Wanted']>=10000]
df = df[df['Worked']>5000]   
df 
```

```python
index = list(df.index)
bar_width = 0.45
pos_worked = np.arange(len(index))
pos_wanted = pos_worked + bar_width
fig = plt.figure(figsize=(19,16))
fig.set_facecolor('#E3E7F0')
plt.barh(pos_worked,df['Worked'],height=bar_width,label='Desiderato',linewidth=1,color='#ead5c1',alpha=0.6)
plt.barh(pos_wanted,df['Wanted'],height=bar_width,label='Utilizzato',linewidth=1,color='#b8e7c7',alpha=0.6)
plt.yticks(pos_worked + bar_width / 2, df.index, fontsize = 17)
plt.title('Le risposte dei Developer su Stack Overflow', fontsize = 25)
plt.legend(bbox_to_anchor=(1, 1), fontsize = 20)
plt.ylabel('Linguaggi di Programmazione', fontsize = 20)
plt.xlabel('\nFrequenza nelle Risposte', fontsize = 20)
plt.xticks(fontsize = 15)
plt.show()
```
