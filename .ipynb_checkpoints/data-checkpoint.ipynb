```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('languages.csv')
df.head()
```

```python
type_freq = pd.crosstab(index = df['type'],
                        columns = ['abs_freq'],
                        colnames = [''],
                        normalize= True)
type_freq
```

```python
type_freq.plot.bar()
plt.show()
```

```python
type_freq.plot.barh(rot = 0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dati di esempio
dati = np.random.randn(1000)

# Crea l'istogramma
n, bins, patches = plt.hist(dati, bins=5, alpha=0.7)

# Colori personalizzati per ciascun bin
colori = ['cornflowerblue', 'royalblue', 'lightsteelblue', 'deepskyblue', 'dodgerblue']

# Assegna colori specifici a ciascun bin
for patch, colore in zip(patches, colori):
    patch.set_facecolor(colore)

# Aggiungi eventuali titoli, etichette, legende, etc.

# Mostra l'istogramma
plt.show()
```

```python
bins = np.arange(1950,2023,3)
l = list()
colori = [(0.6, 0.8, 1.0), (0.5, 0.7, 0.9), (0.4, 0.6, 0.8), (0.3, 0.5, 0.7), (0.2, 0.4, 0.6)]
n,bins, patche = plt.hist(df['appeared'], bins = bins)
for patch, colore in zip(patches,l):
    patch.set_facecolor(l)
plt.grid(alpha = 0.3)

plt.xlabel('Intervallo temporale')
plt.ylabel('Nuovi linguaggi')
plt.show()
```
