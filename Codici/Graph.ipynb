```python
import plotly.graph_objects as go
import pandas as pd

df = pd.read_csv('C:/Users/aless/OneDrive/Desktop/APPUNTI_UNI/visualizzazione_scientifica/CSV/languagesofstates.csv')

x = [[0, 'rgb(0,0,255)'] ,[1, 'rgb(255,0,0)']]

colorscale = ["#E0C3F8","#AED8F6","khaki"]

fig = go.Figure(data=go.Choropleth(
    marker_opacity=0.6,
    locations=df['Code'], # Spatial coordinates
    z = df['Value'].astype(float), # Data to be color-coded
    locationmode = 'USA-states', # set of locations match entries in `locations`
    colorscale = colorscale, #'Reds',
    #colorbar_title = "Millions USD",
))

fig.update_layout(
    autosize = False,
    width = 1000,
    height = 1000,
    title_text = 'Programming languages for state',
    geo_scope='usa', # limite map scope to USA
     geo=dict(
        showcoastlines=True,  # Mostra le linee costiere
        coastlinecolor='black',  # Colore della linea di contorno
        coastlinewidth=1.5,  # Larghezza della linea di contorno
        bgcolor="#E3E7F0"  # Imposta il colore di sfondo
    ),
)

labels = ["python", "JavaScript", "Java"]


fig.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Fetta A', 'Fetta B']
sizes = [40, 60]  # Assicurati che la somma delle dimensioni sia 100%

colors = ['lightcoral', 'lightblue']

explode = (0.1, 0)

circle_border_color = 'black'

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

centre_circle = plt.Circle((0, 0), 0.70, fc='white', ec=circle_border_color, lw=2)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal')

plt.show()
```
