```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pyVSjs = pd.read_csv('C:\Users\saver\OneDrive\Desktop\project\SciVis\Dataset')

#grafico a linee con marker = 'o'"Estrazione dati csv.ipynb"
python_values = pyVSjs['Python']
js_values = pyVSjs['JavaScript']
py_years_values = {}
js_years_values = {}

for i in pyVSjs.index:
    date = pyVSjs['Week'][i]
    date_splitted = date.split('-')
    year = date_splitted[0]

    #py
    if year in py_years_values:
        py_years_values[year].append(pyVSjs['Python'][i])
    else:
        py_years_values[year] = []

    #js
    if year in js_years_values:
        js_years_values[year].append(pyVSjs['JavaScript'][i])
    else:
        js_years_values[year] = []

py_mean = []
for year in py_years_values:
    py_mean.append(np.sum(py_years_values[year])/len(py_years_values[year]))

js_mean = []
for year in js_years_values:
    js_mean.append(np.sum(js_years_values[year])/len(js_years_values[year]))

plt.figure(figsize=(15,10))
plt.plot(py_years_values.keys(), py_mean, linestyle='-', color='r',label='Python',marker='o')
plt.plot(js_years_values.keys(), js_mean, linestyle='-', color='b',label='JavaScript',marker='o')
plt.legend(fontsize=15)
plt.title('Google trends: Python vs Javascript')
plt.xlabel('\nAnni\n')
plt.ylabel('\nRicerche annuali\n')
plt.grid(alpha=0.3)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Carica i dati dal CSV
pyVSjs = pd.read_csv('google-trends-2004-2020.csv')

# Estrai i valori di Python e JavaScript
python_values = pyVSjs['Python']
js_values = pyVSjs['JavaScript']

# Estrai gli anni
years = [int(date.split('-')[0]) for date in pyVSjs['Week']]

# Calcola le medie annuali per Python e JavaScript
py_mean = python_values.groupby(years).mean()
js_mean = js_values.groupby(years).mean()

# Crea il plot delle curve
plt.figure(figsize=(15, 10))
plt.plot(py_mean.index, py_mean, linestyle='-', color='r', label='Python')
plt.plot(js_mean.index, js_mean, linestyle='-', color='b', label='JavaScript')

# Colora l'area sotto le curve
plt.fill_between(py_mean.index, py_mean, color='red', alpha=0.3, label='Area sotto Python')
plt.fill_between(js_mean.index, js_mean, color='blue', alpha=0.3, label='Area sotto JavaScript')

# Aggiungi legenda e dettagli del grafico
plt.legend(fontsize=15)
plt.title('Google trends: Python vs JavaScript')
plt.xlabel('\nAnni\n')
plt.ylabel('\nRicerche annuali\n')
plt.grid(alpha=0.3)

# Mostra il grafico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Carica i dati dal CSV
pyVSjs = pd.read_csv('google-trends-2004-2020.csv')

# Estrai i valori di Python e JavaScript
python_values = pyVSjs['Python']
js_values = pyVSjs['JavaScript']

# Estrai gli anni e i mesi
years, months = zip(*(date.split('-') for date in pyVSjs['Week']))
years = [int(year) for year in years]
months = [int(month) for month in months]

# Calcola le medie mensili per Python e JavaScript
py_mean = python_values.groupby([years, months]).mean()
js_mean = js_values.groupby([years, months]).mean()

# Converte l'indice da tupla a un formato pi maneggevole
js_mean.index = pd.to_datetime(['{}-{}'.format(year, month) for year, month in js_mean.index])

# Calcola array di medie usando NumPy
py_mean_values = py_mean.values
js_mean_values = js_mean.values

plt.figure(figsize=(15,10))
# Crea il plot della curva JavaScript
plt.plot(js_mean.index, js_mean_values, color='b')

plt.plot(js_mean.index, py_mean_values, color='r')

# Aggiungi fill_between solo quando JavaScript  maggiore di Python
plt.fill_between(js_mean.index, js_mean_values, where=(js_mean_values >= py_mean_values), color='blue', alpha=0.3, label='Area sotto JavaScript')

# Aggiungi fill_between solo quando Python  maggiore di JavaScript
plt.fill_between(js_mean.index, py_mean_values, where=(py_mean_values >= js_mean_values), color='red', alpha=0.3, label='Area sotto Python')

# Aggiungi legenda e dettagli del grafico
plt.legend(fontsize=15)
plt.title('Google trends: Python vs JavaScript')
plt.xlabel('\nAnni e Mesi\n')
plt.ylabel('\nRicerche mensili\n')
plt.grid(alpha=0.3)

# Mostra il grafico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Carica i dati dal CSV
pyVSjs = pd.read_csv('google-trends-2004-2020.csv')

# Estrai i valori di Python e JavaScript
python_values = pyVSjs['Python']
js_values = pyVSjs['JavaScript']

# Estrai gli anni e i mesi
years, months = zip(*(date.split('-') for date in pyVSjs['Week']))
years = [int(year) for year in years]
months = [int(month) for month in months]

# Calcola le medie mensili per Python e JavaScript
py_mean = python_values.groupby([years, months]).mean()
js_mean = js_values.groupby([years, months]).mean()

# Converte l'indice da tupla a un formato pi maneggevole
js_mean.index = pd.to_datetime(['{}-{}'.format(year, month) for year, month in js_mean.index])

# Limita il riempimento di JavaScript fino al 2016-01
mask_js = js_mean.index <= '2016-01-01'

# Calcola array di medie usando NumPy
py_mean_values = py_mean.values
js_mean_values = js_mean.values

# Crea il plot delle curve senza fill_between
plt.plot(js_mean.index, js_mean_values, linestyle='-', color='b', label='JavaScript')
plt.plot(js_mean.index, py_mean_values, linestyle='-', color='r', label='Python')

# Trova i punti in cui JavaScript  maggiore di Python
mask = js_mean_values > py_mean_values

# Riempi l'area sopra la curva Python solo dove JavaScript  maggiore
plt.fill_between(js_mean.index, js_mean_values, py_mean_values, where=mask & mask_js, color='blue', alpha=0.3, label='Area sotto JavaScript')

# Aggiungi legenda e dettagli del grafico
plt.legend(fontsize=15)
plt.title('Google trends: Python vs JavaScript')
plt.xlabel('\nAnni e Mesi\n')
plt.ylabel('\nRicerche mensili\n')
plt.grid(alpha=0.3)

# Mostra il grafico
plt.show()
```
